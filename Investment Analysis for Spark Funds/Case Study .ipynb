# Checkpoint 4

````python
mapping = pd.read_csv("./mapping.csv")
mapping.info()
````

````python
transform_mapping =pd.melt(mapping, id_vars=[mapping.columns[0]], value_vars=mapping.columns[1:])
````

````python
category_mapping =transform_mapping.loc[transform_mapping.value==1]
````

````python
category_mapping= category_mapping.drop(['value'],axis=1)
````

````python
category_mapping.rename(columns={'variable':'main sector'}, inplace = True)
category_mapping.rename(columns={'category_list':'primary sector'}, inplace = True)
````

````python
def getPrimarySector(category_list):
    primary_sector=str(category_list).split('|')
    return primary_sector[0]
````

````python
master_frame['primary sector']=master_frame['category_list'].apply(getPrimarySector)
````

````python
master_frame
````

````python
final_master = master_frame.merge(category_mapping,on='primary sector')
````

````python
def get_investment_count_d1(main_sector):
    return  d1.loc[d1['main sector']== main_sector].raised_amount_usd.count()   

def get_investment_sum_d1(main_sector):
    return  d1.loc[d1['main sector']== main_sector].raised_amount_usd.sum()

def get_investment_count_d2(main_sector):
    return  d2.loc[d2['main sector']== main_sector].raised_amount_usd.count()   

def get_investment_sum_d2(main_sector):
    return  d2.loc[d2['main sector']== main_sector].raised_amount_usd.sum()

def get_investment_count_d3(main_sector):
    return  d3.loc[d3['main sector']== main_sector].raised_amount_usd.count()   

def get_investment_sum_d3(main_sector):
    return  d3.loc[d3['main sector']== main_sector].raised_amount_usd.sum()
````

````python
d1['total_number_investments']= d1['main sector'].apply(get_investment_count_d1)
d1['total_sum_investments']= d1['main sector'].apply(get_investment_sum_d1)
````
